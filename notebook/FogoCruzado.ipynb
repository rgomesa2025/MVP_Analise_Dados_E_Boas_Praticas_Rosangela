# Descrição do Problema

O Fogo Cruzado é uma instituição privada sem fins lucrativos que atua em prol de interesses públicos e sociais, contando com a colaboração dos cidadãos para monitorar a violência armada no Brasil.

Sua atuação foca no enfrentamento à escassez e à falta de transparência dos dados oficiais que, historicamente, priorizam apenas crimes consumados, como **homicídios** e **latrocínios**, negligenciando a dinâmica cotidiana dos tiroteios.

Essa lacuna gera um grave déficit democrático, pois o Estado falha em registrar incidentes que, mesmo sem óbitos imediatos, desestruturam a vida urbana, impedem a mobilidade segura da população e silenciam impactos sociais severos, como o fechamento recorrente de escolas e postos de saúde. Ao operar de forma independente do governo, a organização consegue auditar e dar visibilidade a dados que, muitas vezes, são omitidos ou ignorados pelas estatísticas oficiais.

## Hipóteses do Problema

As hipóteses que tracei são as seguintes:

1. **Existe uma concentração estatística da violência armada na faixa etária correspondente aos jovens adultos?**

2. **A violência armada manifesta-se de forma estatisticamente concentrada em polos de alta densidade urbana?**

3. **A ocorrência de tiroteios segue um padrão temporal não aleatório, apresentando ciclos de repetição que coincidem com a rotina de mobilidade urbana?**

## Tipo de Problema

Isso ocorre porque a base de dados é estruturada com rótulos claros (como desfecho da vítima e motivação do crime), permitindo que modelos identifiquem padrões e classifiquem novos eventos com base no histórico de ocorrências monitoradas.

### **Importação e Consolidação dos dados**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

# 1. Definição da URL base do repositório github
GITHUB_RAW_URL = "https://raw.githubusercontent.com/rgomesa2025/MVP_Analise_Dados_E_Boas_Praticas_Rosangela/refs/heads/main/data/"

# 2. Nomes dos arquivos
arquivos_base = [
    'Rio_de_Janeiro/fc_api_occurrences_with_victims_detailed_2026-03-21T17_33_54.000Z.csv',
    'Bahia/fc_api_occurrences_with_victims_detailed_2026-03-21T17_34_10.000Z.csv',
    'Bahia/fc_api_occurrences_with_victims_detailed_2026-03-21T17_34_34.000Z.csv',
    'Pernambuco/fc_api_occurrences_with_victims_detailed_2026-03-21T17_34_25.000Z.csv'
]

def load_from_github(base_url, file_list):
    temp_list = []
    for f in file_list:
        full_url = base_url + f
        try:
            # O pandas consegue ler diretamente de uma URL
            df = pd.read_csv(full_url)
            temp_list.append(df)
            print(f"Sucesso ao ler do GitHub: {f}")
        except Exception as e:
            print(f"ERRO ao ler {f}: {e}")
            print(f"Verifique se a URL está correta: {full_url}")

    if not temp_list:
        raise ValueError("Nenhum arquivo pôde ser carregado. Verifique as URLs.")

    return pd.concat(temp_list, ignore_index=True)

# 3. Execução do Carregamento Remoto
df_completo = load_from_github(GITHUB_RAW_URL, arquivos_base)

# Visualização dos Resultados
print("\n--- Inspeção Inicial do Dataset Consolidado ---")
display(df_completo.head())

print(f"\nTotal de registros consolidados: {df_completo.shape[0]}")
print(f"Total de atributos: {df_completo.shape[1]}")
```

### Análise da Média (Mean)

A média aritmética funciona como o ponto central de equilíbrio dos nossos dados. Para a idade, ela nos indica a faixa etária média das pessoas atingidas, enquanto para a latitude e longitude, ela define as coordenadas centrais das ocorrências. Esta métrica é o primeiro passo para entendermos o "comportamento padrão" dos incidentes em cada estado.

```python
#Tabela de Médias (Formatada)
tabela_media = df_completo[['age', 'latitude', 'longitude']].mean().to_frame().T
tabela_media.columns = ['Média Idade', 'Média Latitude', 'Média Longitude']
print("--- Tabela de Médias Aritméticas ---")
display(tabela_media.style.format("{:.2f}"))
```

```python
# Gráfico: Média de Idade por Estado
plt.figure(figsize=(10, 5))
sns.barplot(x='state', y='age', data=df_completo, estimator='mean', errorbar=None,
            hue='state', palette='mako', legend=False)

plt.title('Média de Idade das Vítimas por Estado', fontsize=14)
plt.ylabel('Idade Média')
plt.xlabel('Estado')
plt.show()
```

### Análise do Desvio Padrão (Std)

O desvio padrão mede a dispersão, ou seja, o quanto os dados se afastam da média. Um valor elevado em 'age' indica que a violência atinge pessoas de idades muito variadas, enquanto nas coordenadas geográficas (latitude e longitude), ele revela se os incidentes estão espalhados por todo o estado ou concentrados em regiões específicas.

```python
# Tabela de Desvio Padrão
tabela_std = df_completo[['age', 'latitude', 'longitude']].std().to_frame().T
tabela_std.columns = ['Desvio Idade', 'Desvio Latitude', 'Desvio Longitude']
print("--- Tabela de Desvio Padrão (Dispersão) ---")
display(tabela_std.style.format("{:.2f}"))
```

```python
# Gráfico: Dispersão da Idade por Estado
plt.figure(figsize=(10, 5))
sns.pointplot(x='state', y='age', data=df_completo, linestyle="none",
              capsize=.2, hue='state', palette='dark:darkorange', legend=False)

plt.title('Dispersão (Desvio Padrão) da Idade por Estado', fontsize=14)
plt.ylabel('Idade (Média e Dispersão)')
plt.xlabel('Estado')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## Preparação do Modelo: Estratégia de Alvos Paralelos

A fim de assegurar que o modelo de aprendizado de máquina seja confiável e livre de vícios nos dados (**overfitting**), aplica-se uma estratégia de modelagem em paralelo. Em vez da predição de um desfecho único, a análise é dividida em duas dimensões críticas, conferindo maior robustez ao estudo:

1.  **Dimensão Operacional (`type`):** Focada na natureza da ocorrência (Ex: presença de vítimas, óbitos ou feridos).
2.  **Dimensão de Cenário (`situation`):** Focada no ambiente onde o evento ocorreu (Ex: via pública, interior de residência ou estabelecimento comercial).

### **Estruturação das Variáveis**
* **Features (X):** Selecionam-se as variáveis preditoras `age`, `latitude` e `longitude` como os dados base para explicar o evento e fundamentar o aprendizado do algoritmo.
* **Target (y):** Definem-se dois alvos distintos (`type` e `situation`) para possibilitar a compreensão de como a localização e a idade influenciam cada desfecho de forma independente.

### **Técnicas de Divisão Aplicadas**
* **Holdout (70/30):** O dataset é particionado em **Treino (70%)**, para que o modelo identifique os padrões, e **Teste (30%)**, onde o algoritmo é desafiado com dados inéditos para a mensuração de sua eficácia real.
* **Estratificação (`stratify=y`):** Esta técnica é utilizada para garantir que a proporção de cada categoria seja preservada integralmente tanto no treino quanto no teste. O procedimento evita que o modelo seja treinado com amostras desequilibradas, assegurando uma avaliação justa de performance.



### **Justificativa da Separação em Dois Modelos**
A opção por dois modelos distintos fundamenta-se no fato de que cada alvo possui uma distribuição estatística própria. Ao isolá-los, garante-se que o aprendizado para o "Tipo" de evento não sofra interferência dos padrões da "Situação". Tal abordagem permite identificar se fatores como localização ou idade apresentam pesos de importância distintos dependendo da ótica analisada.

```python

# Definição de Features (X)
X = df_corr[['age', 'latitude', 'longitude']]

# Divisão para o Modelo 1: Dimensão Operacional (type)
y_type = df_corr['type']
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
    X, y_type, test_size=0.3, random_state=42, stratify=y_type
)

# Divisão para o Modelo 2: Dimensão de Cenário (situation)
y_sit = df_corr['situation']
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X, y_sit, test_size=0.3, random_state=42, stratify=y_sit
)

# Conferência das dimensões resultantes
print("Divisão concluída com sucesso!")
print(f"Modelo Operacional (Type) - Treino: {X_train_t.shape[0]} | Teste: {X_test_t.shape[0]}")
print(f"Modelo de Cenário (Situation) - Treino: {X_train_s.shape[0]} | Teste: {X_test_s.shape[0]}")
```

# Normalização

A etapa de **Normalização** (ou redimensionamento) consiste em transpor os valores das variáveis preditoras `age`, `latitude` e `longitude` para uma escala numérica comum, geralmente compreendida no intervalo entre 0 e 1.

Este procedimento é indispensável devido à sensibilidade de diversos algoritmos de aprendizado de máquina em relação à magnitude e amplitude dos dados. Visto que a variável `age` apresenta valores absolutos consideravelmente superiores às coordenadas geográficas (`latitude` e `longitude`), a ausência de um escalonamento adequado poderia induzir o modelo a atribuir um peso matemático desproporcional à idade, gerando um viés algorítmico.

Portanto, a normalização assegura que todos os atributos contribuam equitativamente para a identificação de padrões e para a construção de modelos preditivos mais precisos e equilibrados.

### **Metodologia Aplicada**
Para a execução desta tarefa, utiliza-se o método `MinMaxScaler`, que realiza a transformação linear dos dados baseando-se nos valores mínimo e máximo de cada atributo, conforme a formulação matemática:

$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$


### **Prevenção de Vazamento de Dados (Data Leakage)**
Como medida de rigor metodológico, o ajuste do normalizador (*fit*) deve ser realizado exclusivamente sobre o **conjunto de treinamento**. A transformação (*transform*) é aplicada aos dados de teste de forma isolada, utilizando os parâmetros obtidos no treino. Esta prática é imprescindível para evitar o vazamento de dados (*data leakage*), garantindo que informações estatísticas do conjunto de teste não influenciem o processo de aprendizado e preservando a capacidade de generalização do modelo final.

```python

# Inicialização do escalonador
scaler = MinMaxScaler()

# --- Normalização do Modelo 1 (Natureza do Evento - Type) ---
X_train_t_scaled = scaler.fit_transform(X_train_t)
X_test_t_scaled = scaler.transform(X_test_t)

# --- Normalização do Modelo 2 (Cenário do Evento - Situation) ---
X_train_s_scaled = scaler.fit_transform(X_train_s)
X_test_s_scaled = scaler.transform(X_test_s)

print("Normalização concluída.")
# --- Exibição das 5 primeiras linhas (Validação) ---
# Convertendo o array do Modelo 1 de volta para DataFrame para visualização clara
df_train_t_preview = pd.DataFrame(X_train_t_scaled, columns=X.columns)

print("\n--- Primeiras 5 linhas dos dados normalizados (Modelo Type) ---")
display(df_train_t_preview.head())

# Verificação estatística rápida
print("\nEstatísticas do conjunto normalizado:")
display(df_train_t_preview.describe().loc[['min', 'max']])
```

```python
# Configuração do estilo visual
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Título Geral
fig.suptitle('Distribuição das Variáveis Preditoras Após Normalização (Min-Max Scaling)', fontsize=16)

# Lista de colunas e cores para o loop
columns = ['age', 'latitude', 'longitude']
colors = ['#4e79a7', '#f28e2b', '#e15759']

# Gerar os gráficos de distribuição para cada variável
for i, col in enumerate(columns):
    sns.histplot(df_train_t_preview[col], kde=True, ax=axes[i], color=colors[i], bins=30)
    axes[i].set_title(f'Distribuição: {col.capitalize()}')
    axes[i].set_xlabel('Valor Normalizado (0-1)')
    axes[i].set_ylabel('Frequência')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Perspectivas para Evolução do Pré-processamento

Embora a normalização e a limpeza de dados forneçam uma base consistente, a engenharia de dados avançada oferece ferramentas para extrair padrões mais profundos. Abaixo, descrevem-se as etapas complementares que poderiam ser integradas para otimizar a performance preditiva:

### **1. Engenharia de Características (Feature Engineering)**
Em vez de utilizar apenas os dados brutos, pode-se criar novas variáveis que capturem o contexto do evento. No conjunto de dados em questão, duas transformações seriam altamente relevantes:
* **Agrupamento Temporal:** Extrair o "Turno do Dia" (Manhã, Tarde, Noite) a partir da coluna de data/hora, visto que a dinâmica da violência urbana varia conforme o horário.
* **Variáveis de Proximidade:** Calcular a distância dos eventos em relação a pontos de interesse (escolas, hospitais ou delegacias) utilizando as colunas de `latitude` e `longitude`.

### **2. Redução de Dimensionalidade (PCA - Principal Component Analysis)**
O PCA é uma técnica estatística que transforma um conjunto de variáveis correlacionadas em um número menor de variáveis não correlacionadas, chamadas de componentes principais.
* **Aplicação:** Caso o dataset venha a incluir dezenas de colunas socioeconômicas da região, o PCA ajudaria a reduzir a complexidade do modelo, retendo a maior parte da variância dos dados e eliminando ruídos, o que previne o *overfitting*.



### **3. Seleção de Características (Feature Selection)**
Nem toda informação disponível é útil para a predição. Utilizar métodos como **Recursive Feature Elimination (RFE)** ou análise de **Feature Importance** permitiria descartar colunas que não contribuem estatisticamente para prever o `type` ou a `situation`. Isso torna o modelo mais leve, rápido e fácil de interpretar.

### **4. Codificação de Variáveis Categóricas (Encoding)**
Para que variáveis como `region` ou `neighborhood` sejam utilizadas no modelo, seria necessário aplicar técnicas de **One-Hot Encoding** ou **Target Encoding**. Isso converteria nomes de bairros em vetores numéricos, permitindo que o algoritmo compreenda padrões geográficos além das coordenadas decimais.

A identificação dessas "manchas de calor" temporais prova a existência de uma sazonalidade diária e semanal, permitindo concluir que a violência armada possui um comportamento previsível e diretamente vinculado à dinâmica de funcionamento das metrópoles.
