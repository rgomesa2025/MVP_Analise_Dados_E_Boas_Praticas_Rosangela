# Descrição do Problema

O Fogo Cruzado é uma instituição privada sem fins lucrativos que atua em prol de interesses públicos e sociais, contando com a colaboração dos cidadãos para monitorar a violência armada no Brasil.

Sua atuação foca no enfrentamento à escassez e à falta de transparência dos dados oficiais que, historicamente, priorizam apenas crimes consumados, como **homicídios** e **latrocínios**, negligenciando a dinâmica cotidiana dos tiroteios.

Essa lacuna gera um grave déficit democrático, pois o Estado falha em registrar incidentes que, mesmo sem óbitos imediatos, desestruturam a vida urbana, impedem a mobilidade segura da população e silenciam impactos sociais severos, como o fechamento recorrente de escolas e postos de saúde. Ao operar de forma independente do governo, a organização consegue auditar e dar visibilidade a dados que, muitas vezes, são omitidos ou ignorados pelas estatísticas oficiais.
