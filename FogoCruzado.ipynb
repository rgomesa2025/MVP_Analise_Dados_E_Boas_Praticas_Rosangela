# Descrição do Problema

O Fogo Cruzado é uma instituição privada sem fins lucrativos que atua em prol de interesses públicos e sociais, contando com a colaboração dos cidadãos para monitorar a violência armada no Brasil.

Sua atuação foca no enfrentamento à escassez e à falta de transparência dos dados oficiais que, historicamente, priorizam apenas crimes consumados, como **homicídios** e **latrocínios**, negligenciando a dinâmica cotidiana dos tiroteios.

Essa lacuna gera um grave déficit democrático, pois o Estado falha em registrar incidentes que, mesmo sem óbitos imediatos, desestruturam a vida urbana, impedem a mobilidade segura da população e silenciam impactos sociais severos, como o fechamento recorrente de escolas e postos de saúde. Ao operar de forma independente do governo, a organização consegue auditar e dar visibilidade a dados que, muitas vezes, são omitidos ou ignorados pelas estatísticas oficiais.

## Hipóteses do Problema

As hipóteses que tracei são as seguintes:

1. **A violência armada impacta o acesso a direitos básicos de forma desproporcional?**

2. **Existe correlação direta entre operações policiais e o aumento de balas perdidas em áreas residenciais?**

3. **A violência armada segue um padrão temporal previsível ligado à rotina urbana?**

## Tipo de Problema

Sob a ótica da ciência de dados, esse problema enquadra-se no **Aprendizado Supervisionado**, especificamente em tarefas de Classificação. Isso ocorre porque a base de dados é estruturada com rótulos claros (como desfecho da vítima e motivação do crime), permitindo que modelos identifiquem padrões e classifiquem novos eventos com base no histórico de ocorrências monitoradas.

## Seleção de Dados

Para a execução deste MPV, o dataset do Fogo Cruzado oferece um conjunto de dados que podem ser processados através de bibliotecas de aprendizado de máquina amplamente utilizadas em Python, como Pandas para manipulação, Scikit-Learn para modelagem e partição de treino/teste e XGBoost ou Random Forest, ideais para lidar com a complexidade das variáveis categóricas presentes.

Embora o conjunto de dados apresente um alto nível de curadoria técnica, ele exige **etapas fundamentais de pré-processamento antes da integração aos modelos**. A seleção e o tratamento de dados são indispensáveis para gerenciar valores ausentes em colunas demográficas e realizar a engenharia de recursos o "feature engineering". Esse processo é necessário para converter o contexto social e as descrições qualitativas em vetores numéricos processáveis, garantindo que os dados, embora já confiáveis e estruturados, tornem-se aptos para algoritmos de classificação eficazes.
